# Sección 2 Lección 9: Implementando la visualización de datos con matplotlib

```python
import pandas as pd

# Datos en formato diccionario
datos = {
    "Nombre": ["Lucía", "Carlos", "María", "David", "Ana", "Pepe", "Pepe"],
    "Edad": [28, 35, 22, 41, 30, 18, 18],
    "Sexo": ["F", "M", "F", "M", "F", "M", "M"],
    "Ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla", "Bilbao", "Badajoz", "Badajoz"],
    "Salario": [32000, 45000, 29000, 51000, 38000, 27000, 27000]
}

# Crear DataFrame
df = pd.DataFrame(datos)
```

```python
# Guardar como CSV
df.to_csv("datos.csv", index=False)
```

```python
df = pd.read_csv("datos.csv")
print(df)
```

```python
# Revisar duplicados y eliminarlos si hay
print(f'Duplicados: {df.duplicated().sum()}')
df.drop_duplicates(inplace=True)
```

```python
print(df)
```

## Gráfico de barras verticales

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))       # Crea una figura de tamaño 8x5 pulgadas
plt.bar(df['Nombre'],            # Usa la columna 'Nombre' como etiquetas del eje X
        df['Salario'],           # Usa la columna 'Salario' como alturas de las barras
        color='skyblue')         # Color de las barras: azul claro
plt.title('Salario por persona') # Título del gráfico
plt.xlabel('Nombre')             # Etiqueta del eje X
plt.ylabel('Salario (€)')        # Etiqueta del eje Y
plt.xticks(rotation=45)          # Rota las etiquetas del eje X 45 grados para mejor visualización
plt.tight_layout()               # Ajusta automáticamente el espaciado para que no se solapen los elementos
plt.show()                       # Muestra el gráfico
```

## Gráfico de líneas

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Edad'],        # Dibuja un gráfico de líneas con: # - Edad en el eje X
         df['Salario'],     # - Salario en el eje Y
         marker='o',        # - Puntos marcados con círculos (marker='o')
         linestyle='-',     # - Línea continua (linestyle='-')
         color='green')     # - Color verde
plt.title('Relación entre Edad y Salario')  
plt.xlabel('Edad')  
plt.ylabel('Salario (€)')  
plt.grid(True)                  # Activa una cuadrícula de fondo para facilitar la lectura
plt.tight_layout()  
plt.show()
#NO ES APLICABLE EN ESTOS DATOS, NO SON SECUENCIALES EN EL TIEMPO
```

## Gráfica de dispersión o puntos

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Edad'], df['Salario'], color='purple')
plt.title('Edad vs Salario')
plt.xlabel('Edad')
plt.ylabel('Salario (€)')
plt.tight_layout()
plt.show()
```

```python
sexo_counts = df['Sexo'].value_counts() # Cuenta cuántas veces aparece cada valor en la columna 'Sexo' (por ejemplo, cuántos hombres y cuántas mujeres)
print(sexo_counts)

plt.figure(figsize=(5, 5))
plt.pie(  
    sexo_counts,                         # Valores a representar en el gráfico circular (número de personas por sexo)
    labels=sexo_counts.index,            # Etiquetas para cada segmento (por ejemplo: 'F' y 'M')
    autopct='%1.1f%%',                   # Muestra el porcentaje con un decimal en cada segmento
    startangle=90,                       # Rota el gráfico para que empiece desde arriba (90 grados)
    colors=['pink', 'lightblue']         # Colores asignados a cada categoría
)
plt.title('Distribución por sexo')  
plt.axis('equal')                        # Asegura que el gráfico tenga forma de círculo perfecto (relación 1:1 entre los ejes)
plt.show()  
```

# Carga de dataset de clasificación

```python
#Cargar dataset
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")
```

```python
"""
survived -->	Si el pasajero sobrevivió (1) o no (0).
pclass -->	    Clase del pasajero: 1 = Primera clase, 2 = Segunda clase, 3 = Tercera clase.
sex -->	        Sexo del pasajero (male = hombre, female = mujer).
age -->	        Edad del pasajero en años. Puede tener valores faltantes (NaN).
sibsp -->	    Número de hermanos o cónyuges a bordo (Siblings/Spouses).
parch -->	    Número de padres o hijos a bordo (Parents/Children).
fare -->	    Tarifa pagada por el pasajero para el billete.
embarked -->	Puerto de embarque: C = Cherbourg, Q = Queenstown, S = Southampton.
class -->	    Clase en formato texto (First, Second, Third), equivalente a pclass pero en texto.
who -->	        Clasificación simplificada de la persona: man, woman, child.
adult_male -->	Booleano: True si es un hombre adulto, False en caso contrario.
deck -->	    Cubierta del barco donde estaba el pasajero (A, B, C, etc.).
embark_town -->	Nombre completo del puerto de embarque (Cherbourg, Queenstown, Southampton).
alive -->	    Texto que indica si sobrevivió o no ("yes" o "no").
alone -->	    Booleano que indica si el pasajero estaba solo (True) o acompañado (False) en el barco.
"""
```

```python
#Vista inicial
df
```

```python
# 1. Revisar info y valores nulos
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
# 2. Imputar o eliminar valores nulos
# Edad (age) tiene valores nulos, imputamos con la mediana
df['age'].fillna(df['age'].median(), inplace=True) #inplace es para sustituir directamente en el mismo objeto df
```

```python
print(df.isnull().sum())
```

```python
# Embarked tiene pocos nulos, imputamos con la moda
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True) #Ponemos el primer valor de la serie moda
```

```python
print(df.isnull().sum())
```

```python
# 3. Eliminar columnas poco útiles, redundantes o con muchos valores nulos
df.drop(columns=['deck', 'alive', 'class', 'who', 'adult_male', 'embark_town', 'alone'], inplace=True)
```

```python
df
```

```python
# 4. Convertir variables categóricas a numéricas con LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ['sex', 'embarked']:
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
# Contar cuántos pasajeros embarcaron en cada puerto (0=Cherbourg, 1=Queenstown, 2=Southampton)
embarked_counts = df['embarked'].value_counts().sort_index()  # Cuenta las ocurrencias y ordena por índice

# Crear un gráfico de barras con los nombres de los puertos en el eje X y el número de pasajeros en el eje Y
plt.bar(['Cherbourg', 'Queenstown', 'Southampton'], embarked_counts, color='teal')  # Dibuja las barras con etiquetas y color

plt.title('Pasajeros por puerto de embarque')
plt.ylabel('Número de pasajeros')
plt.show()
```

```python
plt.hist(df['age'].dropna(), bins=10, color='purple')  # Histograma de la columna 'age' sin valores nulos, con 30 intervalos (bins) y color morado
#plt.hist(df['age'].dropna(), bins=[0, 10, 20, 40, 60, 80], color='purple') #Ancho eje x personalizado
plt.title('Distribución de edades')
plt.xlabel('Edad')
plt.ylabel('Número de pasajeros')
plt.show()
```

```python
survival_by_class = pd.crosstab(df['pclass'], df['survived'])  # Crea una tabla de contingencia entre clase (pclass) y supervivencia (survived)


survival_by_class.plot(kind='bar', stacked=True, color=['red', 'green'], figsize=(8,6))  
# Crea un gráfico de barras apiladas:
# kind='bar' → tipo de gráfico: barras verticales
# stacked=True → apila las barras para ver proporciones en cada clase
# color=['red', 'green'] → rojo para los que murieron (0), verde para los que sobrevivieron (1)
# figsize=(8,6) → tamaño de la figura en pulgadas
plt.title('Supervivencia por clase')
plt.xlabel('Clase')
plt.ylabel('Número de pasajeros')
plt.legend(['Murió', 'Sobrevivió'])
plt.show()
```

```python
#BARRAS CON ERROR
categories = ['Clase 1', 'Clase 2', 'Clase 3']  # Nombres de las categorías (etiquetas del eje X)
means = [20, 35, 30]  # Valores medios para cada categoría
errors = [2, 3, 4]  # Errores estándar o márgenes de error correspondientes a cada media


plt.bar(categories, means, yerr=errors, capsize=5, color='skyblue')  
# Crea un gráfico de barras:
# categories → etiquetas en el eje X
# means → altura de cada barra
# yerr=errors → añade barras de error (líneas verticales) a cada barra
# capsize=5 → tamaño de los extremos horizontales de las barras de error
# color='skyblue' → color de las barras
plt.title('Barras con barras de error')
plt.ylabel('Valor medio')
plt.show()
```

```python
import numpy as np
x = np.random.rand(50)  # Genera 50 valores aleatorios entre 0 y 1 para el eje X
y = np.random.rand(50)  # Genera 50 valores aleatorios entre 0 y 1 para el eje Y
sizes = 100 * np.random.rand(50)  # Genera 50 tamaños aleatorios para los puntos (de 0 a 100)
colors = np.random.rand(50)       # Genera 50 colores aleatorios (valores entre 0 y 1)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')  
# Crea un diagrama de dispersión (scatter plot):
# x, y → coordenadas de los puntos
# s=sizes → tamaño de cada punto
# c=colors → color de cada punto según el valor
# alpha=0.6 → transparencia (1 = opaco, 0 = transparente)
# cmap='viridis' → paleta de colores usada para los puntos
plt.colorbar()  # Muestra una barra lateral que representa la escala de colores
plt.title('Scatter plot con tamaño y color variables')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

## Librería Seaborn basada en Matplotlib

```python
import seaborn as sns
sns.countplot(data=df, x='sex', hue='survived', palette='Set1')  
# data=df → usa el DataFrame df como fuente de datos  
# x='sex' → el eje X representa la variable 'sex' (sexo: 0 = mujer, 1 = hombre)  
# hue='survived' → divide cada barra por la variable 'survived' (0 = murió, 1 = sobrevivió), creando subgrupos por color  
# palette='Set1' → usa una paleta de colores predefinida llamada 'Set1' para diferenciar las categorías del hue
plt.xlabel('Sexo (0=F, 1=M)')  # Etiqueta del eje X
plt.ylabel('Número de pasajeros')  # Etiqueta del eje Y
plt.title('Supervivencia por sexo')  # Título del gráfico
plt.legend(['Murió', 'Sobrevivió'])  # Leyenda manual para indicar qué color representa qué
plt.show()  # Mostrar el gráfico
```

```python
#MATRIZ DE CORRELACIÓN
plt.figure(figsize=(10, 6))
corr = df.corr()             # Calcular la matriz de correlación entre las variables numéricas del DataFrame
sns.heatmap(
    corr,                    # Datos: matriz de correlación
    annot=True,              # Mostrar los valores numéricos en cada celda
    cmap='coolwarm',         # Paleta de colores que va de tonos fríos (azul) a cálidos (rojo)
    fmt=".2f"                # Formato para mostrar los valores con 2 decimales
)
plt.title('Mapa de correlación entre variables numéricas')
plt.show()
```

```python
#HISTOGRAMA CON CURVA KDE(SUAVIZA LA DISTRIBUCIÓN)
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,                  # Dataset con los datos
    x='age',                  # Variable numérica que se va a graficar (edad)
    hue='survived',           # Dividir el histograma según supervivencia (0=no, 1=sí)
    kde=True,                 # Añadir curva KDE (estimación de densidad) para suavizar la distribución
    multiple='stack',         # Apilar las barras del histograma según la variable hue
    palette=['red', 'green']  # Colores para los grupos de supervivencia: rojo=murió, verde=sobrevivió
)
plt.title('Distribución de edad según supervivencia')
plt.xlabel('Edad')
plt.ylabel('Cantidad de pasajeros')
plt.show()
```

```python
#BOXPLOT
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,                  # Dataset con los datos
    x='pclass',               # Variable categórica en eje X: clase del pasajero (1, 2, 3)
    y='fare',                 # Variable numérica en eje Y: precio del billete
    hue='sex',                # División adicional por sexo (0=F, 1=M) para comparar dentro de cada clase
    palette='Set2'            # Paleta de colores Set2 para distinguir los grupos por sexo
)
plt.title('Distribución del precio del billete por clase y sexo')
plt.xlabel('Clase')
plt.ylabel('Precio del billete')
plt.show()
```

```python
#GRÁFICO DE VIOLÍN
plt.figure(figsize=(10, 6))
sns.violinplot(
    data=df,                 # Dataset que contiene los datos
    x='sex',                 # Variable categórica en el eje X (sexo: 0=F, 1=M)
    y='age',                 # Variable numérica en el eje Y (edad)
    hue='survived',          # División adicional dentro de cada categoría por supervivencia (0=no, 1=sí)
    split=True              # Divide cada violín en dos mitades según la variable hue (survived)
)
plt.title('Distribución de edad por sexo y supervivencia')
plt.xlabel('Sexo (0=F, 1=M)')
plt.ylabel('Edad')
plt.show()
```

```python
#MATRIZ DE CONFUSIÓN
survival_pivot = df.pivot_table(
    index='pclass',          # Las filas serán las clases de pasajero (1, 2, 3)
    columns='sex',           # Las columnas serán el sexo (0=Femenino, 1=Masculino)
    values='survived',       # Los valores que queremos agregar son los de la columna 'survived'
    aggfunc='mean'           # Usamos la media para obtener la tasa de supervivencia
)


sns.heatmap(
    survival_pivot,          # Datos para el heatmap (tabla creada antes)
    annot=True,              # Muestra los valores numéricos dentro de cada celda
    cmap='YlGnBu',           # Paleta de colores amarillos-verdes-azules para el mapa de calor
    fmt=".2f"                # Formato de los números: dos decimales
)

plt.title('Tasa de supervivencia por clase y sexo')
plt.xlabel('Sexo (0=F, 1=M)')
plt.ylabel('Clase')
plt.show()
```

```python
#GRÁFICO CON SUBGRÁFICOS
x = np.linspace(0, 2*np.pi, 400)  # Crea 400 valores entre 0 y 2π para el eje X
y = np.sin(x**2)  # Calcula sin(x²) para cada valor de x

fig, axs = plt.subplots(2, 2, figsize=(10,8))  
# Crea una figura con una cuadrícula de 2x2 subplots (axs), tamaño 10x8 pulgadas
# fig es el contenedor general de los gráficos, axs es un array 2x2 de ejes (gráficos individuales)

axs[0, 0].plot(x, y, 'tab:blue')  # Primer subplot: gráfico de línea azul de y = sin(x²)
axs[0, 0].set_title('Plot 1')     # Título del primer subplot

axs[0, 1].scatter(x, y, color='tab:orange')  # Segundo subplot: diagrama de dispersión en naranja
axs[0, 1].set_title('Plot 2')                # Título del segundo subplot

axs[1, 0].bar(['A', 'B', 'C'], [3, 7, 4], color='tab:green')  # Tercer subplot: gráfico de barras
axs[1, 0].set_title('Plot 3')                                # Título del tercer subplot

axs[1, 1].hist(np.random.randn(1000), bins=30, color='tab:red')  # Cuarto subplot: histograma de 1000 valores aleatorios (distribución normal)
axs[1, 1].set_title('Plot 4')                                    # Título del cuarto subplot

plt.tight_layout()
plt.show()
```
