```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
import numpy as np

# Cargar dataset
autos = fetch_openml(name='autoPrice', version=1, as_frame=True, parser='auto')

X = autos.data.select_dtypes(include=[np.number])
y = autos.target.astype(float)
# Limpiar valores nulos poniendo la media
X = X.fillna(X.mean())

# División
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, root_mean_squared_error

# Mismo preprocesamiento anterior
model_tree = DecisionTreeRegressor(max_depth=5, random_state=42)
model_tree.fit(X_train, y_train)
y_pred_tree = model_tree.predict(X_test)
print(f"R²: {r2_score(y_test, y_pred_tree):.3f}")
print(f"RMSE: {root_mean_squared_error(y_test, y_pred_tree):.2f}")
```
